```python
import pandas as pd

orders_df = pd.read_csv("orders.csv")
print("ORDERS DATA")
print(orders_df)
```

```python
users_df = pd.read_json("users.json")
print("\nUSERS DATA")
print(users_df)
```

```python
import sqlite3

conn = sqlite3.connect("restaurants.db")

with open("restaurants.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)

restaurants_df = pd.read_sql("SELECT * FROM restaurants", conn)

print("\nRESTAURANTS DATA")
print(restaurants_df)
```

```python
orders_users_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)

print("\nORDERS + USERS")
print(orders_users_df)
```

```python
final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)

print("\nFINAL DATASET")
print(final_df)
```
